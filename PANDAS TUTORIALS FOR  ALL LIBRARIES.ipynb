# drop null values:

```python
df1 = pd.DataFrame({"name":['Hatmin','Jasmine','Shaktiman'],
                   "toy":[np.nan,"football","piano"],
                   "born":[pd.NaT,pd.Timestamp("2000-05-23"),
                          pd.NaT]})
```

```python
df1
```

```python
df1.dropna()
```

# drop duplicates

```python
df2 = pd.DataFrame({"name":['Hatmin','Jasmine','Jasmine'],
                   "toy":[np.nan,"football","piano"],
                   "born":[pd.NaT,pd.Timestamp("2000-05-23"),
                          pd.NaT]})
```

```python
df2
```

```python
df2.drop_duplicates(subset="name")
```

```python
df2.drop_duplicates(subset="name",keep ="last")   #it will print 2nd duplicate;
```

```python
df2.drop_duplicates(subset="name",keep =False)   #it will not print any duplicates; 
```
